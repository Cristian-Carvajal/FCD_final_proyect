### <span style="color:#2F749F;"><strong>Ejercicio 1: C치lculo de medidas de tendencia central</strong></span>

A partir de una muestra simulada de ingresos, calcula la **media, mediana y moda**, y luego a interpretarlas.

1. Usa el n칰mero base `544038` (debes reemplazarlo con los 칰ltimos 6 d칤gitos de tu documento de identificaci칩n).  
2. Con este n칰mero genera una semilla (`np.random.seed`) para que los resultados sean reproducibles.  
3. Los ingresos se simulan con una distribuci칩n normal:  
   - Media te칩rica: `(base % 10) + 2`  
   - Desviaci칩n est치ndar: `7`  
   - Tama침o de muestra: `150`  
4. Los valores negativos se reemplazan por cero (pues los ingresos no pueden ser negativos).  
5. Grafica tus resultados (ver imagen de ejemplo)

<p align="center">
    <img src="imgs\sp_2_e1.png" alt="sp_2_e1" width="900"  height="400">
</p>


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns

prueba = 544038
np.random.seed(prueba)

data = np.random.normal(loc=(prueba % 10) + 2, scale=7, size=150).round(0)

media = np.mean(data)
mediana = np.median(data)
moda = stats.mode(data, keepdims=True)[0][0]

print(f"Media:   {media:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Moda:    {moda:.2f}")

# Visualizaci칩n
plt.boxplot(data, vert=False)
plt.title("Edad de las personas que ingresaron a un parque de diversiones")
plt.xlabel("Cantidad de personas")
plt.ylabel("Variable: Edad")
plt.axvline(media, color='red', linestyle='--', label=f'Media: {media:.2f}')
plt.axvline(mediana, color='green', linestyle='-', label=f'Mediana: {mediana:.2f}')
plt.axvline(moda, color='blue', linestyle=':', label=f'Moda: {moda:.2f}')
plt.legend()
plt.show()
```

```python
# Visualizaci칩n
plt.figure(figsize=(10, 5))
sns.histplot(data, bins=10, kde=False, color="#2F749F")

# L칤neas de tendencia central
plt.axvline(media, color='red', linestyle='--', linewidth=2, label=f'Media: {media:.2f}')
plt.axvline(mediana, color='green', linestyle='-', linewidth=2, label=f'Mediana: {mediana:.2f}')
plt.axvline(moda, color='purple', linestyle=':', linewidth=2, label=f'Moda: {moda:.2f}')

# Est칠tica

plt.title("Edad de las personas que ingresaron a un parque de diversiones")
plt.xlabel("Edad")
plt.ylabel("Cantidad de personas")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### <span style="color:#2F749F;"><strong>Ejercicio 2: C치lculo de medidas de dispersi칩n</strong></span>

A partir de la muestra generada en el Ejercicio 1, analiza qu칠 tan dispersos est치n los datos respecto a su media. Calcula:
1. Rango
2. Varianza
3. Desviaci칩n est치ndar
4. Coeficiente de variaci칩n
5. Genera una gr치fica con los resultados obtenidos (ver imagen de ejemplo)

<p align="center">
    <img src="imgs\sp_2_e2.png" alt="sp_2_e2" width="900"  height="400">
</p>

```python
# C치lculo de medidas de dispersi칩n
media = np.mean(data)
minimo = np.min(data)
maximo = np.max(data)
rango = maximo - minimo
varianza = np.var(data, ddof=1)
desviacion = np.std(data, ddof=1)
coef_var = desviacion / media

print(f"Minimo:      {minimo:.2f}")
print(f"Maximo:      {maximo:.2f}")
print(f"rango:       {rango:.2f}")
print(f"varianza:    {varianza:.2f}")
print(f"desviacion:  {desviacion:.2f}")
print(f"coef_var:    {coef_var:.2f}")

# Visualizaci칩n
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(data, bins=6, kde=True, color='skyblue')

# L칤neas verticales
plt.axvline(media, color='green', linestyle='--', linewidth=2, label=f'Media = {media:.2f}')
plt.axvline(media + desviacion, color='purple', linestyle=':', linewidth=1.5, label=f'+1픢 = {media + desviacion:.2f}')
plt.axvline(media - desviacion, color='purple', linestyle=':', linewidth=1.5, label=f'-1픢 = {media - desviacion:.2f}')
plt.axvline(minimo, color='red', linestyle='-', linewidth=1.5, label=f'M칤nimo = {minimo}')
plt.axvline(maximo, color='red', linestyle='-', linewidth=1.5, label=f'M치ximo = {maximo}')

# T칤tulo y etiquetas
plt.title("Edad de las personas que ingresaron a un parque de diversiones")
plt.xlabel("Edad")
plt.ylabel("Cantidad de personas")
plt.legend()


# Recuadro con medidas
texto = (
    f"Rango: {rango}\n"
    f"Varianza: {varianza:.2f}\n"
    f"Desviaci칩n est치ndar: {desviacion:.2f}\n"
    f"Coef. de variaci칩n: {coef_var:.2f}"
)
plt.text(7.5, 3, texto, bbox=dict(facecolor='lightgreen', alpha=0.3), fontsize=10)

plt.tight_layout()
plt.show()
```

#### Medidas de dispersi칩n

Podemos decir que las edades de las personas que visitan el parque de atracciones var칤an en promedio 6 a침os con respecto a la edad media de este mismo grupo (10 a침os)

```python
from scipy.stats import skew, kurtosis
import pandas as pd

# Estilo visual
sns.set_theme(style="whitegrid")

# Simulaci칩n de datos
data
df = pd.DataFrame({'Datos': data})

# C치lculo de m칠tricas
asimetria = skew(df['Datos'])
curtosis_valor = kurtosis(df['Datos'])

# Visualizaci칩n
plt.figure(figsize=(10, 6))
sns.histplot(df['Datos'], kde=True, color='skyblue', bins=15)
plt.axvline(df['Datos'].mean(), color='red', linestyle='--', label='Media')

# Anotaciones en la esquina superior izquierda
plt.text(0.05, 0.95, f"Asimetr칤a: {asimetria:.2f}", transform=plt.gca().transAxes,
         fontsize=12, verticalalignment='top', bbox=dict(facecolor='white', alpha=0.6))
plt.text(0.05, 0.88, f"Curtosis: {curtosis_valor:.2f}", transform=plt.gca().transAxes,
         fontsize=12, verticalalignment='top', bbox=dict(facecolor='white', alpha=0.6))

# T칤tulos y etiquetas
plt.title("Edad de las personas que ingresaron a un parque de diversiones")
plt.xlabel("Edad")
plt.ylabel("Cantidad de personas")
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
#np.percentile(data, 25)
# Estilo visual
sns.set_theme(style="whitegrid")

# C치lculo de percentiles clave
percentiles = {
    'P10': {'valor': np.percentile(data, 10), 'color': 'red', 'linestyle': '-'},
    'P25 (Q1)': {'valor': np.percentile(data, 25), 'color': 'orange', 'linestyle': '--'},
    'P50 (Mediana)': {'valor': np.percentile(data, 50), 'color': 'green', 'linestyle': '-.'},
    'P75 (Q3)': {'valor': np.percentile(data, 75), 'color': 'blue', 'linestyle': ':'},
    'P90': {'valor': np.percentile(data, 90), 'color': 'purple', 'linestyle': (0, (5, 2))}
}

# C치lculo de deciles (D1 a D9)
deciles = {f'D{i}': np.percentile(data, i * 10) for i in range(1, 10)}

# Visualizaci칩n
plt.figure(figsize=(14, 7))
sns.histplot(data, kde=True, color='skyblue', bins=20)

# L칤neas verticales para percentiles destacados
for etiqueta, props in percentiles.items():
    plt.axvline(props['valor'], color=props['color'], linestyle=props['linestyle'],
                linewidth=2)
    plt.text(props['valor'], plt.ylim()[1]*0.9, etiqueta,
             rotation=90, color=props['color'], fontsize=10,
             verticalalignment='center', horizontalalignment='right',
             backgroundcolor='white')

# L칤neas m치s sutiles para deciles
for etiqueta, valor in deciles.items():
    plt.axvline(valor, color='gray', linestyle='--', alpha=0.3)
    plt.text(valor, plt.ylim()[1]*0.6, etiqueta,
             rotation=90, color='gray', fontsize=8,
             verticalalignment='center', horizontalalignment='right')

# Recuadro interpretativo
texto = (
    "Percentiles dividen en 100 partes\n"
    "Deciles dividen en 10 partes (D1 a D9)\n"
    "Cuantiles como Q1, Q2, Q3 son fracciones (0.25, 0.50, 0.75)\n"
    "칔tiles para analizar posici칩n relativa y desigualdad"
)
plt.text(0.98, 0.95, texto, transform=plt.gca().transAxes,
         fontsize=10, verticalalignment='top', horizontalalignment='right',
         bbox=dict(boxstyle="round,pad=0.5", facecolor="white", edgecolor="gray"))

# T칤tulos y estilo
plt.title("Distribuci칩n Normal con Percentiles y Deciles", fontsize=14, fontweight='bold')
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.tight_layout()
plt.show()
```

En esta segunda parte, genera **dos variables** a partir de t칰 n칰mero de c칠dula y realiza un an치lisis **bivariado** para calcular la relaci칩n entre ellas.

1. Use los **6 칰ltimos d칤gitos** de su c칠dula como base (similar al Ejercicio 1).  
2. Genere dos variables aleatorias (ejemplo: `ingresos` y `gastos`) con distribuci칩n normal, usando esa base como **semilla reproducible**.  
3. Calcule los coeficientes de correlaci칩n de **Pearson** y **Spearman**.
4. Grafique las variables analizadas (ver la imagen de ejemplo)
5. Interprete los resultados.  

<p align="center">
    <img src="imgs\sp_2_e5.png" alt="sp_2_e5" width="900"  height="400">
</p>

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

# Caso 1: Relaci칩n lineal
x1 = np.linspace(0, 7, 150)
y1 = 2 * x1 + np.random.normal(loc=(7 % 10) + 2, scale=7, size=150).round(0)

# Correlaciones
def calcular_correlaciones(x, y):
    pearson, _ = stats.pearsonr(x, y)
    spearman, _ = stats.spearmanr(x, y)
    return pearson, spearman

p1, s1 = calcular_correlaciones(x1, y1)

# Visualizaci칩n
fig, ax = plt.subplots(figsize=(8, 5))  # Solo una subgr치fica
ax.scatter(x1, y1, color="blue", alpha=0.7)
ax.set_title("Relaci칩n Lineal")
ax.set_xlabel("X")
ax.set_ylabel("Y")

# Agregar texto con las correlaciones
texto1 = f"Pearson: {p1:.2f}\nSpearman: {s1:.2f}"
ax.text(0.05, 0.95, texto1, transform=ax.transAxes,
        fontsize=11, verticalalignment='top',
        bbox=dict(boxstyle="round", facecolor="white", edgecolor="gray"))

plt.show()
```
