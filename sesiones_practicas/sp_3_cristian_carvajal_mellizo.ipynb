1. Carga el conjunto de datos penguins desde la librer칤a seaborn `load_dataset("penguins")`.

   - 'bill_length_mm'

   - 'body_mass_g'

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

df = sns.load_dataset("penguins")
df.head()
```

```python
df2 = df.drop(["island", "sex"], axis=1)
df2
```

```python
print(len(df2))
df2.info()
```

```python
df2 = df2.dropna()
df2.info()
```

```python
columns = df2.columns
columns = columns.drop('species')
columns
```

```python
# Estandarizar variables num칠ricas
scaler = StandardScaler()
X_std = scaler.fit_transform(df2[columns])
```

```python
# PCA
pca = PCA()
X_pca = pca.fit_transform(X_std)
```

```python
sns.pairplot(df, hue='species', palette='Set2', diag_kind='hist', corner=False)
plt.suptitle('Matriz de dispersi칩n: relaciones entre variables del Iris', fontsize=16, y=1.02)
plt.show()
```

```python
# Matriz de componentes
componentes = pd.DataFrame(pca.components_.T,
                           columns=[f'PC{i+1}' for i in range(len(columns))],
                           index=columns)

# 游늻 Visualizaci칩n de autovectores
plt.figure(figsize=(8, 5))
sns.heatmap(componentes, annot=True, cmap='coolwarm', center=0)
plt.title('Autovectores')
plt.ylabel('Variables originales')
plt.xlabel('Componentes principales')
plt.tight_layout()
plt.show()

# Ecuaciones de los componentes principales con nombres reales
print("Ecuaciones de los componentes principales:\n")
for i, comp in enumerate(componentes.columns):
    pesos = pca.components_[i]
    ecuacion = " + ".join([f"{round(pesos[j], 3)}췅'{columns[j]}'" for j in range(len(pesos))])
    print(f"{comp} = {ecuacion}")
```

```python
# Varianza explicada
varianza = pca.explained_variance_
PVE = pca.explained_variance_ratio_
PVE_acum = np.cumsum(PVE)
componentes = [f'PC{i+1}' for i in range(len(PVE))]

# DataFrame para visualizaci칩n
df_varianza = pd.DataFrame({
    'Componente': componentes,
    'Autovalor': varianza,
    'PVE': PVE,
    'PVE acumulada': PVE_acum
})

# Proporci칩n de varianza explicada
plt.figure(figsize=(8, 5))
sns.barplot(x='Componente', y='PVE', data=df_varianza, color='skyblue', label='PVE individual')
sns.lineplot(x='Componente', y='PVE acumulada', data=df_varianza, marker='o', color='darkblue', label='PVE acumulada')
plt.axhline(0.7, color='red', linestyle='--', label='Umbral 70%')
plt.axhline(0.9, color='green', linestyle='--', label='Umbral 90%')
plt.title('Proporci칩n de varianza explicada por componente')
plt.ylabel('Proporci칩n de varianza')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
df_pca = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(len(PVE))])
correlaciones = df_pca.corr()


df_pca['species'] = df['species']

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_pca, x='PC1', y='PC2', hue='species', palette='Set2')
plt.title('Proyecci칩n de los datos en los dos primeros componentes')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy.stats import shapiro
from scipy.special import kolmogorov
from statsmodels.graphics.gofplots import qqplot
from sklearn.model_selection import train_test_split

# Cargar datos
diamonds = sns.load_dataset("diamonds")
diamonds = diamonds.dropna()

# Mostrar primeras filas
diamonds.head()
```

```python
diamonds.info()
```

```python
# Visualizacion de una muestra aleatoria de datos ya que pairplot genera problemas de rendimiento con muchos datos
diamonds_sample = diamonds.sample(n=2000, random_state=42)

sns.pairplot(diamonds_sample, hue='cut', palette='Set2', diag_kind='hist', corner=False)
plt.suptitle('Matriz de dispersi칩n: relaciones entre variables del diamante', fontsize=16, y=1.02)
plt.show()
```

#### Validaci칩n de supuestos

Tras graficar una peque침a representaci칩n de los datos presentes en el dataset 띿iamonds, podemos verificar como algunas de las variables presentan una relaci칩n lineal, adem치s, podemos ver como las variables x, y, z presentan una tendencia exponencial con respecto a la variable precio.
Finalmente, podemos decir que la variable carat presenta una ligera tendencia lineal con respecto a la variable precio


```python
X = diamonds[['carat', 'depth', 'table', 'x', 'y', 'z']]
Y = diamonds['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size=0.30, random_state=42)
```

```python
diamonds_clean = diamonds.copy()
diamonds_clean = diamonds_clean.drop(["cut", "color", "clarity"], axis=1)
diamonds_clean.info()
```

```python
# Agregar constante para el intercepto
X_const = sm.add_constant(X_train)

# Ajustar modelo
modelo = sm.OLS(y_train, X_const).fit()

resid = modelo.resid
# Resumen estad칤stico
print(modelo.summary())
```
